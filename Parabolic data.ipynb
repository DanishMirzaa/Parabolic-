```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
import random
Y=[]
a=random.choice(np.arange(0.6,2,0.1))
k=random.choice(np.arange(5))
h=random.choice(np.arange(5))
sign=random.choice([-1,1])
for X in np.arange(-5,5,0.1):    
    var=random.choice(np.arange(-200,250)) 
    
    Y1=sign*((a*(X-h)**2)+var)
    Y.append(Y1)
plt.plot(Y)   
```

```python
X=np.ones([1,2,200])
Y=[1]
for i in range(10):
    y=[]
    a=random.choice(np.arange(0.6,2,0.1))
    k=random.choice(np.arange(5))
    h=random.choice(np.arange(-2,2,0.5))
    sign=random.choice([-1,1])
    Y.append(sign)
    
    for x in  np.arange(-10,10,0.1):
        var=random.choice(np.arange(-25,25))
        var2=random.choice(np.arange(-25,25))
        y1=sign*((a*(x-h)**2)+k+var)
        y2=sign*((a*(x-h)**2)+k+var2)
        Y1=np.
        y.append(y1)
    X=np.vstack([X,np.array(y).reshape(1,200)])    
    
Y=np.array(Y)
X=X[1:,:]
Y=Y[1:]
```

```python
Y=tf.one_hot(Y,2)
Y
```

```python
model=tf.keras.Sequential()
model.add(tf.keras.layers.Dense(10,activation='relu'))
model.add(tf.keras.layers.Dense(10,activation='relu'))
model.add(tf.keras.layers.Dense(10,activation='relu'))
model.add(tf.keras.layers.Dense(10,activation='relu'))

model.add(tf.keras.layers.Dense(2,activation='softmax'))
```

```python
model.compile(loss=tf.keras.losses.CategoricalCrossentropy(),optimizer=tf.keras.optimizers.Adam(lr=0.001),metrics='accuracy')
```

```python
model.fit(X,Y,epochs=1)
```

## evaluate model 

```python
X_test=np.ones([1,200])
Y_test=[1]
for i in range(500):
    y=[]
    a=random.choice(np.arange(10,20,0.1))
    k=random.choice(np.arange(5))
    h=random.choice(np.arange(-2,2,0.5))
    sign=random.choice([-1,1])
    Y_test.append(sign)
    for x in  np.arange(-10,10,0.1):
        var=random.choice(np.arange(-500,500))
        y1=sign*((a*(x-h)**2)+k+var)
        y.append(y1)
    X_test=np.vstack([X_test,np.array(y).reshape(1,200)])    
    
Y_test=np.array(Y_test)
X_test=X_test[1:,:]
Y_test=Y_test[1:]
Y_test=tf.one_hot(Y_test,2)
model.evaluate(X_test,Y_test)
```
